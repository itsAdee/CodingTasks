```python
def linear_regression_predictions_for_vectors_vectors(weight_vector, feature_vector):
    if len(weight_vector) != len(feature_vector):
        raise ValueError("The weight and feature vectors must be the same length.")

    prediction = 0
    for weight, feature in zip(weight_vector, feature_vector):
        prediction += weight * feature

    return prediction


# Testing the function
# Define a weight vector and a feature vector
weight_vector = [0.5, 1.2, -0.3,3.4]
feature_vector = [1,2.0, -1.0, 3.0]

# Use the function to make a prediction
prediction = linear_regression_predictions_for_vectors_vectors(weight_vector, feature_vector)

print("The prediction is:", prediction)
```

```python
import numpy as np

def linear_regression_predictions_for_matrix_vectors(weight_vector, feature_matrix):
    return np.dot(feature_matrix, weight_vector)

# Define a weight vector and a feature matrix
weight_vector = np.array([0.5, 1.2, -0.3])
feature_matrix = np.array([[2.0, -1.0, 3.0], [1.0, 2.0, -1.0], [-1.0, 1.0, 2.0]])

# Use the function to make predictions
predictions = linear_regression_predictions_for_matrix_vectors(weight_vector, feature_matrix)

print("The predictions are:", predictions)
```

```python
import numpy as np

def mean_squared_error(predictions, actuals):
    return np.mean((predictions - actuals) ** 2)

# Define a vector of predictions and a vector of actual values
predictions = np.array([0.5, 1.2, -0.3])
actuals = np.array([0.6, 1.1, -0.2])

# Use the function to calculate the Mean Squared Error
mse = mean_squared_error(predictions, actuals)

print("The Mean Squared Error is:", mse)
```

```python
import numpy as np

def mse_gradient(predictions, actuals):
    return 2 * np.mean(predictions - actuals)

# Define a vector of predictions and a vector of actual values
predictions = np.array([0.5, 1.2, -0.3])
actuals = np.array([0.6, 1.1, -0.2])

# Use the function to calculate the gradient of the Mean Squared Error
gradient = mse_gradient(predictions, actuals)

print("The gradient of the Mean Squared Error is:", gradient)
```

```python
def vanilla_gradient_descent(weight_vector, actuals, gradient_of_loss, stopping_criterion, feature_matrix, learning_rate=0.001, max_iterations=10000):
    for _ in range(max_iterations):
        gradient = gradient_of_loss(weight_vector, actuals, feature_matrix)
        weight_vector -= learning_rate * gradient
        if stopping_criterion(weight_vector, gradient):
            return weight_vector
    return weight_vector

def stochastic_gradient_descent(weight_vector, actuals, gradient_of_loss, stopping_criterion, feature_matrix, learning_rate=0.001, max_iterations=10000):
    num_samples = feature_matrix.shape[0]
    for _ in range(max_iterations):
        for i in range(num_samples):
            gradient = gradient_of_loss(weight_vector, actuals[i], feature_matrix[i:i+1])
            weight_vector -= learning_rate * gradient
            if stopping_criterion(weight_vector, gradient):
                return weight_vector
    return weight_vector

# Define a weight vector
weight_vector = np.zeros(X_normalized.shape[1])

# Use the function to perform vanilla gradient descent
optimized_weights_vanilla = vanilla_gradient_descent(weight_vector, y, gradient_of_loss, stopping_criterion, X_normalized)

print("The optimized weights (vanilla) are:", optimized_weights_vanilla)

# Use the function to perform stochastic gradient descent
optimized_weights_stochastic = stochastic_gradient_descent(weight_vector, y, gradient_of_loss, stopping_criterion, X_normalized)

print("The optimized weights (stochastic) are:", optimized_weights_stochastic)
```
